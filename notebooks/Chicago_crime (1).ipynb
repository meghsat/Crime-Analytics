```python
from kmodes.kmodes import KModes

kclusters = 20
# df_new=df.drop(['Location','Date','Updated On','Location Description'],axis=1)
# df_new.drop(df_new.loc[df_new['Primary Type']=='OTHER WEAPONS VIOLATION'].index, inplace=True)
# run k-means clustering
kmeans = KModes(n_clusters=kclusters, random_state=42).fit(df_2016_1000)
# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:201]
```

```python
from geopy.geocoders import Nominatim

address = 'Chicago, Il'

geolocator = Nominatim(user_agent="bg_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Chicago City are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_bnglr = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, block in zip(df_2016_1000['Latitude'], df_2016_1000['Longitude'], df_2016_1000['Block']):
    label = '{}'.format(block)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_bnglr)  
    
map_bnglr
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df_2016_1000['Latitude'], df_2016_1000['Longitude'],df_2016_1000['Block'], df_2016_1000['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[int(cluster)-1],
        fill=True,
        fill_color=rainbow[int(cluster)-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
